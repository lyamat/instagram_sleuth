```python
def get_post_liked_users(posts_or_tagged: str):
    href_value = driver.current_url
    match = re.search(r'/p/([^/]+)/', href_value)

    if match:
        extracted_portion = match.group(1)
        base_url = "https://www.instagram.com/p/"
        post_href = f"{base_url}{extracted_portion}/"
            
    if check_element("//a[contains(@href, 'liked_by')]"):
        liked_by_button = driver.find_element(By.XPATH, "//a[contains(@href, 'liked_by')]")
        
        
        post_liked_users = set()
        
        if posts_or_tagged == "posts":
            liked_by_button.click()
            
            WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH,"//h1[@class='_ac78']")))

            height = driver.find_element(By.XPATH, "//div[contains(@style, '356')]//div[contains(@style, 'column')]").value_of_css_property("padding-top")
            # match = False
            sleep(1.5)
            try:
                while True:
                    lastHeight = height

                    elements = driver.find_elements(By.XPATH, ".//span[@class='_ap3a _aaco _aacw _aacx _aad7 _aade']")

                    for element in elements:
                        if element.text not in post_liked_users:
                            post_liked_users.add(element.text)

                    driver.execute_script("return arguments[0].scrollIntoView();", elements[-1])
                    

                    height = driver.find_element(By.XPATH, "//div[contains(@style, '356')]//div[contains(@style, 'column')]").value_of_css_property("padding-top")
                    if lastHeight==height:
                        break
            except Exception as e:
                print(f"Some exception: {e}")
                
            close_div = driver.find_element(By.XPATH,"//div[@class='_ac76']")
            close_button = close_div.find_element(By.XPATH, ".//*[local-name()='svg' and @aria-label='Close']")
            close_button.click()

        return post_liked_users, post_href
    return set(), post_href
    
```

```python
def get_post_commented_usernames():
    load_more_comments_button_xpath = "//*[local-name()='svg' and @aria-label='Load more comments']"
    if (check_element(load_more_comments_button_xpath)):
        while True:
            try:
                WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, load_more_comments_button_xpath)))
            except TimeoutException:
                print("All comments are loaded")
                break
            
            load_more_comments_button = driver.find_element(By.XPATH,load_more_comments_button_xpath)
            load_more_comments_button.click()
            
            sleep(1)
            WebDriverWait(driver, 5).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='_aano']//div[1][@data-visualcompletion='loading-state']")))
    
        
    # all_usernames_commented = driver.find_elements(By.XPATH, "//div[@class='x78zum5 xdt5ytf x1iyjqo2']//span[@class='_ap3a _aaco _aacw _aacx _aad7 _aade']")
    try:
        WebDriverWait(driver, 1).until(EC.presence_of_element_located((By.XPATH, "//ul[@class='_a9ym']")))
    except TimeoutException:
        return dict(), dict()
    # комменты основные
    all_usernames_commented = driver.find_elements(By.XPATH,"//ul[@class='_a9ym']//span[@class='xt0psk2']//a")
    usernames_and_count_1 = defaultdict(int)
    for username in all_usernames_commented:
        usernames_and_count_1[username.text] += 1
    
    usernames_and_count_2 = defaultdict(int)
    try:
        WebDriverWait(driver, 1).until(EC.presence_of_element_located((By.XPATH, "//button[@class=' _acan _acao _acas _aj1- _ap30']//div[@class='_a9yh'][1]")))
        # пооткрывать все комменты под комментами
        view_replies_buttons = driver.find_elements(By.XPATH,"//button[@class=' _acan _acao _acas _aj1- _ap30']//div[@class='_a9yh'][1]")
        for button in view_replies_buttons:
            button.click()
            
        sleep(0.5)
        WebDriverWait(driver, 5).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='_aano']//div[1][@data-visualcompletion='loading-state']")))

        # комменты под комментами
        all_replied_usernames_commented = driver.find_elements(By.XPATH,"//ul[@class='_a9z6 _a9za']//ul[@class='_a9ym']//ul[@class='_a9yo']//span[@class='xt0psk2']")
        
        for username in all_replied_usernames_commented:
            usernames_and_count_2[username.text] += 1
            
    except TimeoutException:
        pass
    
    return usernames_and_count_1, usernames_and_count_2
```

```python
def analyze_likes(post_publisher, posts_or_tagged):
    from collections import Counter

    with open(f"{post_publisher}---{posts_or_tagged}.json", "r") as f:
        data = json.load(f)

    likes_counter = Counter()

    for post in data.values():
        likes_counter.update(post['post_liked_users'])

    results = {}

    top_10_users = [user for user, _ in likes_counter.most_common(10)]
    results['top_10_users'] = top_10_users

    user_likes = {user: count for user, count in likes_counter.items()}
    results['user_likes'] = user_likes

    with open(f"{post_publisher}---{posts_or_tagged}_analyze_likes.json", "w") as f:
        json.dump(results, f, indent=2)
```

```python
def scan_posts_or_tagged(posts_or_tagged: str):
    if (posts_or_tagged == "posts"):
        posts_button = driver.find_element(By.XPATH, "//span[text()='Posts']")
        posts_button.click()
    elif (posts_or_tagged == "tagged"):
        tagged_button = driver.find_element(By.XPATH, "//a[contains(@href, 'tagged')]")
        tagged_button.click()
    
    sleep(1)
    divs = driver.find_elements(By.XPATH, "//div[@class='_aabd _aa8k  _al3l']")
    
    posts = {}
    
    pattern = r"https://www\.instagram\.com/([a-zA-Z0-9._]+)"
    match = re.match(pattern, driver.current_url)
    if match:
        post_publisher = match.group(1)

    for div in divs:
            post = {}
            try:
                div.click()
                # если карусель
                if check_element("//div[@class='_aamj _acvz _acnc _acng']"):
                    parent_div = driver.find_element(By.XPATH, "//div[@class='_aamj _acvz _acnc _acng']")
                    child_divs = parent_div.find_elements(By.XPATH, "./div")
                    num_child_divs = len(child_divs)

                    img_index = 1

                    while True:
                        img_xpath = f"//div[@class='_aamm']//li[@class='_acaz'][{img_index}]//img"
                        try:
                            img = driver.find_element(By.XPATH, img_xpath)
                            source = img.get_attribute('src')
                            # img_queue.put(source)
                        except NoSuchElementException:
                            pass

                        try:
                            next_button = driver.find_element(By.XPATH, "//button[@class=' _afxw _al46 _al47']")
                            next_button.click()

                            if img_index < num_child_divs:
                                img_index += 1

                        except NoSuchElementException:
                            break
                        
                else:
                    try:
                        img = div.find_element(By.XPATH, ".//img[@class='x5yr21d xu96u03 x10l6tqk x13vifvy x87ps6o xh8yej3']")
                        source = img.get_attribute('src')
                        # img_queue.put(source)
                    except NoSuchElementException:
                        pass

                # ----------------комменты----------------
                post_usernames_comments, post_replied_usernames_commented_comments = get_post_commented_usernames()
                post['post_usernames_comments'] = [{'username': k, 'count_of_comments': v} for k, v in post_usernames_comments.items()]
                post['post_replied_usernames_commented_comments'] = [{'username': k, 'count_of_comments': v} for k, v in post_replied_usernames_commented_comments.items()]
                # ----------------комменты (конец)----------------
                
                # ----------------Ники лайкеров и ссылка на пост----------------
                post_liked_users, post_href = get_post_liked_users(posts_or_tagged)
                post['post_liked_users'] = list(post_liked_users)
                # ----------------Ники лайкеров и ссылка на пост (конец)----------------
            
                posts[post_href] = post
            
                close_button = driver.find_element(By.XPATH, "//*[local-name()='svg' and @aria-label='Close']")
                close_button.click()

            except NoSuchElementException:
                print("NoSuchElementException in scan_posts_or_tagged")
            
    with open(f"{post_publisher}---{posts_or_tagged}.json", "w") as f:
        json.dump(posts, f, indent=2)
        
    if (posts_or_tagged == "posts"):
        analyze_likes(post_publisher, posts_or_tagged)
```

```python
def get_usernames_follow(who: str) -> set:
    
    cur_url_username = driver.current_url.replace("https://www.instagram.com/", "").rstrip("/")
    who_count = int(WebDriverWait(driver, 3).until(
        EC.presence_of_element_located((By.XPATH, f"//a[contains(@href, '/{who}')]//span//span"))
    ).text)
    
    if who == "followers" and who_count > 300:
        print(f"{cur_url_username} {who} > than 300 ")
        return list()

    elif who == "following" and who_count > 200:
        print(f"{cur_url_username} {who} > than 200 ")
        return list()

    follow_button = driver.find_element(By.XPATH, f"//a[contains(@href, '/{who}')]")
    follow_button.click()

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class='_aano']//div[1][@data-visualcompletion='loading-state']")))
    WebDriverWait(driver, 10).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='_aano']//div[1][@data-visualcompletion='loading-state']")))

    div_follow_list = WebDriverWait(driver, 3).until(
        EC.presence_of_element_located((By.CLASS_NAME, "_aano"))
    )
    
    last_ht, height = 0, 1
    while True:
        last_ht = height
        try:
            height = driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight); return arguments[0].scrollHeight;", div_follow_list)
            sleep(0.2)
            WebDriverWait(driver, 5).until(EC.invisibility_of_element_located((By.XPATH, "//div[@data-visualcompletion='loading-state']")))
            if(height == last_ht):
                break
            
        except TimeoutException:
            break
    
    
    
    links = div_follow_list.find_elements(By.XPATH, "//span[@class='_ap3a _aaco _aacw _aacx _aad7 _aade']")
    
    profile_info = {
        "cur_url_username": cur_url_username,
        f"{who}_count": len(links),
        "followers": [],
        "following": [],
    }

    driver.execute_script("arguments[0].scrollTop = 0;", div_follow_list)
    sleep(0.5)
    
    for link in links:
        ActionChains(driver).move_to_element(link).perform()

        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,"//div[@class='x78zum5 xdt5ytf xg6iff7 x1n2onr6']/div[3]//div//div//div")))
        appear_div = driver.find_element(By.XPATH, "//div[@class='x78zum5 xdt5ytf xg6iff7 x1n2onr6']/div[3]//div//div//div")
        
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,"//div[@class='xexx8yu x1sxyh0 x18d9i69 xurb0ha x2b8uid'][2]//span//span")))
        followers_count = appear_div.find_element(By.XPATH, "//div[@class='xexx8yu x1sxyh0 x18d9i69 xurb0ha x2b8uid'][2]//span//span").text
        
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,"//div[@class='xexx8yu x1sxyh0 x18d9i69 xurb0ha x2b8uid'][3]//span//span")))
        following_count = appear_div.find_element(By.XPATH, "//div[@class='xexx8yu x1sxyh0 x18d9i69 xurb0ha x2b8uid'][3]//span//span").text
        
        is_account_private = check_element("//i[@data-visualcompletion='css-img']")
        
        driver.execute_script("arguments[0].style.left='100px';", appear_div)
        
        follow_info = {
            "username": link.text,
            "followers_count": followers_count,
            "following_count": following_count,
            "is_private": is_account_private,
        }
        
        profile_info[f"{who}"].append(follow_info)
        
    
    close_button = driver.find_element(By.XPATH, "//*[local-name()='svg' and @aria-label='Close']")
    close_button.click()
    
    return profile_info
```
